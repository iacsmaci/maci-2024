```python
# Group by 'City' and calculate mean 'Score' and sum 'Salary'
agg_df = df.group_by('City').agg([
    pl.col('Score').mean().alias('mean_score'),
    pl.col('Salary').sum().alias('total_salary')
])
print("\nAggregated Data:")
display(agg_df)
```

```python
# Fill missing values in 'Salary' with the mean salary
mean_salary = df['Salary'].mean()
filled_df = df.with_columns(pl.col('Salary').fill_null(mean_salary))
print("\nDataFrame with Missing Values Filled:")
display(filled_df)
```

```python
# Filter rows based on a condition
filtered_df = df.filter(df['Age'] > 30)
print("\nFiltered Rows (Age > 30):")
filtered_df.show()
```

### 5. Aggregating Data

```python
# Group by 'City' and calculate mean 'Score' and sum 'Salary'
from pyspark.sql.functions import mean, sum

agg_df = df.groupBy('City').agg(
    mean('Score').alias('mean_score'),
    sum('Salary').alias('total_salary')
)
print("\nAggregated Data:")
agg_df.show()
```

### 6. Dealing with Missing Values

```python
from pyspark.sql.functions import avg

# Fill missing values in 'Salary' with the mean salary
mean_salary = df.select(avg('Salary')).first()[0]
filled_df = df.na.fill({'Salary': mean_salary})
print("\nDataFrame with Missing Values Filled:")
filled_df.show()
```

## Bokeh

```python
!pip install bokeh
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool, CDSView, BooleanFilter
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Prepare the data source for Bokeh
source = ColumnDataSource(df)

# Create a new plot with a title and axis labels
p = figure(title="Iris Dataset: Sepal Length vs Sepal Width", x_axis_label='Sepal Length (cm)', y_axis_label='Sepal Width (cm)')

# Add a hover tool to display info
hover = HoverTool()
hover.tooltips = [("Species", "@species"), ("Sepal Length", "@{sepal length (cm)}"), ("Sepal Width", "@{sepal width (cm)}")]
p.add_tools(hover)

# Add the scatter plot
color_map = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
for species, color in color_map.items():
    subset = source.data['species'] == species
    p.circle('sepal length (cm)', 'sepal width (cm)', size=8, source=source, color=color, legend_label=species, view=CDSView(source=source, filters=[BooleanFilter(subset)]))

# Legend positioning
p.legend.location = "top_left"

# Output the plot to the notebook
output_notebook()
show(p)
```

This project aims to harness the power of big data analytics to enhance the operational efficiency and customer satisfaction of a shared bike system.

Using a comprehensive dataset spanning two years of operations in a major city, this analysis will focus on evaluating the performance of bike rental stations. ### 1. How does bike rental vary across different months?

```python
data_train['datetime'] = pd.to_datetime(data_train['datetime'])
data_train['month'] = data_train['datetime'].dt.month
data_train
```

```python
# Group data by month and calculate average rental count
monthly_rentals = data_train.groupby('month')['count'].mean()

# Plotting monthly rentals
plt.figure(figsize=(10, 5))
monthly_rentals.plot(kind='bar', color='skyblue')
plt.title('Average Bike Rentals by Month')
plt.xlabel('Month')
plt.ylabel('Average Count')
plt.xticks(rotation=0)  # Rotates the x-axis labels to horizontal
plt.show()
```

But what happens if we add more information to the problem and consider other variables for analyzing the average bike rentals by month, such as temperature, humidity, apparent temperature, and wind speed?

```python
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

fs = 18

# Extract month from datetime
data_train['month'] = data_train['datetime'].dt.month

# Group data by month and calculate average rentals, temperature, and humidity
monthly_data = data_train.groupby('month').agg({
    'count': 'mean',        # Average rentals
    'temp': 'mean',         # Average temperature
    'humidity': 'mean'      # Average humidity
}).reset_index()

# Plotting the results in subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6), sharey=True)  # 1 row, 2 columns

# First subplot for Temperature and Bike Rentals
ax1.bar(monthly_data['month'], monthly_data['count'], color='skyblue', label='Bike Rentals')
ax1.set_xlabel('Month', fontsize=fs)
ax1.set_ylabel('Average Bike Rentals', fontsize=fs, color='b')
ax1.tick_params(axis='x', labelsize=fs-2)
ax1.tick_params(axis='y', labelcolor='b', labelsize=fs-2)
ax1.set_xticks(monthly_data['month'])
ax1.set_xticklabels(month_names)
ax1.set_title('Bike Rentals and Temperature', pad=13, size=fs+2)

# Second subplot for Humidity and Bike Rentals
ax2.bar(monthly_data['month'], monthly_data['count'], color='lightgreen', label='Bike Rentals')
ax2.set_xlabel('Month', fontsize=fs)
ax2.set_ylabel('Average Bike Rentals', fontsize=fs, color='b')
ax2.tick_params(axis='x', labelsize=fs-2)
ax2.tick_params(axis='y', labelcolor='b', labelsize=fs-2)
ax2.set_xticks(monthly_data['month'])
ax2.set_xticklabels(month_names)
ax2.set_title('Bike Rentals and Humidity', pad=13, size=fs+2)

# Create a second y-axis for temperature
ax3 = ax1.twinx()
ax3.plot(monthly_data['month'], monthly_data['temp'], color='r', marker='o', label='Temperature (°C)')
ax3.set_ylabel('Temperature (°C)', fontsize=fs, color='r')
ax3.tick_params(axis='y', labelcolor='r', labelsize=fs-2)

# Create a second y-axis for humidity
ax4 = ax2.twinx()
ax4.plot(monthly_data['month'], monthly_data['humidity'], color='g', marker='x', label='Humidity (%)')
ax4.set_ylabel('Humidity (%)', fontsize=fs, color='g')
ax4.tick_params(axis='y', labelcolor='g', labelsize=fs-2)

plt.tight_layout()
plt.show()
```

```python
fs = 18  # Font size for all text elements

data_train['hour'] = data_train['datetime'].dt.hour

# Classify each entry as Holiday, Business Day, or Non-Working Day (weekend or non-working holiday)
data_train['day_type'] = 'Non-Working Day'  # default
data_train.loc[data_train['workingday'] == 1, 'day_type'] = 'Business Day'
data_train.loc[data_train['holiday'] == 1, 'day_type'] = 'Holiday'

# Group data by hour and day type, then calculate the average count of rentals
hourly_type_rentals = data_train.groupby(['day_type', 'hour'])['count'].mean().unstack(level=0)

# Plotting the results
fig, ax = plt.subplots(figsize=(15, 7))
# Using markers for better visualization and differentiating day types with colors
hourly_type_rentals.plot(ax=ax, marker='o', linestyle='-', markersize=8, linewidth=2,
                         color=['royalblue', 'forestgreen', 'crimson'])
plt.title('Hourly Bike Rentals: Business Days, Non-Working Days, and Holidays', fontsize=fs, pad=13)
plt.xlabel('Hour of the Day', fontsize=fs)
plt.ylabel('Average Count of Rentals', fontsize=fs)
plt.legend(title='Day Type', fontsize=fs, title_fontsize=fs)
plt.grid(True)
plt.xticks(range(0, 24), fontsize=fs-2)  # Ensure all hours are shown and adjusted font size
plt.yticks(fontsize=fs-2)
plt.show()
```
